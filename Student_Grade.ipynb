```python
def process_scores(students):
    averages = {}

    for name, scores in students.items():
        avg = round(sum(scores) / len(scores), 2)
        averages[name] = avg

    return averages
students = {}

num_students = int(input("Enter number of students: "))

for _ in range(num_students):
    name = input("Enter student name: ")

    scores = []
    for i in range(3):
        score = int(input(f"Enter score {i+1}: "))
        scores.append(score)

    students[name] = scores


averages = process_scores(students)
print("Averages:", averages)
```

```python
def classify_grades(averages):
    result = {}

    for name, avg in averages.items():
        if avg >= 90:
            grade = 'A'
        elif avg >= 75 and avg < 90:
            grade = 'B'
        elif avg >= 60 and avg < 75 :
            grade = 'C'
        else:
            grade = 'F'

        result[name] = (avg, grade)

    return result


grades = classify_grades(averages)

print("Grades:", grades)
```

```python
def generate_report(classified, passing_avg=60):
    passed = 0

    print("\nStudent Report")
    print("---------------------------")
    print("Name\tAverage\tGrade\tStatus")

    for name, (avg, grade) in classified.items():
        status = "Pass" if avg >= passing_avg else "Fail"
        if status == "Pass":
            passed += 1

        print(f"{name}\t{avg}\t{grade}\t{status}")

    return passed
passed_students = generate_report(grades)

print("\nTotal students passed:", passed_students)
```
